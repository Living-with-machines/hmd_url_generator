```python
from hmd_newspaper_dl.core import get_download_urls, get_newspaper_links
from tqdm.auto import tqdm
import re
import json
from hmd_newspaper_dl.core import _download
```

```python
title_urls = get_newspaper_links()
```

```python
title_urls[0]
```

```python
sample = title_urls[:2]
```

```python
def infer_date_from_fname(fname:str):
    parts = re.split(r"_|\.",fname)
    for p in parts:
        if len(p)==4:
            return p
    else:
        print(fname)
```

```python
title_download_urls = {}
for title, url in tqdm(sample):
    download_urls = get_download_urls(url)
    date_dict = {}
    for url in tqdm(download_urls, leave=False):
        fname = _download(url, return_filename_only=True,sleepy=False)
        date = infer_date_from_fname(fname)
        #{date: {"url": url, "fname":fname}}
        # date_dict['url']= url
        # date_dict['fname'] = fname
        date_dict[date] = {"url": url, "fname":fname}
    title_download_urls[title] = date_dict
```

```python
with open('HMD_title_urls.json', 'w') as f:
    json.dump(title_download_urls, f)
```
